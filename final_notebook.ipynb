```python
def get_details(text):
    split=text.split(":")
    try:
        Start_up_scheme=(split[0]+''.join(split[1].split("\n")[:-1])).strip()
        time_period=(''.join(split[3].split("\n")[:-1])).strip()
        industry=(''.join(split[6].split("\n")[:-1])).strip()
        overview=(''.join(split[7].split("\n")[:-1])).strip()
        fiscal_incentive=(''.join(''.join(split[8:]).split("\n"))).strip()
    except:
        pass
    return {"scheme":Start_up_scheme,"time":time_period,"industry":industry,"overview":overview,"fiscal_incentive":fiscal_incentive}
```
